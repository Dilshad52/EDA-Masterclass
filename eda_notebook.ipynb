```python
###Revenue Histogram
```

```python
import matplotlib.pyplot as plt

df["Revenue"].hist()

plt.title("Revenue Distribution")

plt.show()
```

```python
###Customer Revenue Bar Chart
```

```python
plt.figure(figsize=(8,5))

plt.bar(df["Customer"], df["Revenue"])

plt.title("Revenue by Customer")

plt.xlabel("Customer")

plt.ylabel("Revenue")

plt.show()
```

```python
###Revenue vs Profit Scatter Plot
```

```python
plt.scatter(
    df["Revenue"],
    df["Profit"]
)

plt.xlabel("Revenue")
plt.ylabel("Profit")

plt.title("Revenue vs Profit")

plt.show()
```

```python
###Correlation Matrix
```

```python
df.corr()
```

```python
###Correlation Heatmap
```

```python
import matplotlib.pyplot as plt

corr = df[["Age","Revenue","Cost","Profit"]].corr()

plt.figure(figsize=(6,5))

plt.imshow(corr)

plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)

plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Matrix")

plt.show()
```
